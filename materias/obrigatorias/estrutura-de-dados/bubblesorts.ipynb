```python
from time import time
from random import randint
import matplotlib.pyplot as plt
import numpy as np


time1_list = []
time2_list = []
test_n = []
```

```python
wait = time()
for j in range(500):

    A = [randint(1,1000) for i in range(j+1)]

    A_2 = A

    outOfPlace = -1

    inicio = time()

    tamanho = len(A) - 1

    while ( outOfPlace != 0 ):
        outOfPlace = 0
        for i in range(tamanho):
            i -= 1
            if ( (A[i] < A[i+1]) ):
                aux = A[i + 1]
                A[i + 1] = A[i]
                A[i] = aux
                outOfPlace += 2
        tamanho -= 1
        
    A = [A[-1]] + A[:-1]
    
    fim = time()

    t1 = (fim - inicio)*1000

    tamanho = len(A) - 1
    
    inicio = time()

    for i in range(tamanho):
        for j in range(tamanho-1):
            if ( A_2[i] > A_2[j] ):
                aux = A_2[j]
                A_2[j] = A_2[i]
                A_2[i] = aux   

    fim = time()

    t2 = (fim - inicio)*1000
    
    time1_list.append( t1 )
    time2_list.append( t2 )
    test_n.append(j)
    
for_it = time()

print("O tempo para fazer os dois bubble sorts foi {}".format(for_it - wait))
```

```python
np_t1 = np.array(time1_list)
np_t2 = np.array(time2_list)
np_test_n = np.array(test_n)

fig = plt.figure(figsize = (10,10))

plt.title('Gráfico da diferença entre eficência do Bubble Sort entre os métodos usados') 

plt.xlabel('Número de elementos no vetor')
plt.ylabel('Razão da eficiência entre o método 1 e 2')

plt.plot(np_test_n, np_t1, label = 'Método 1')
plt.plot(np_test_n, np_t2, label = 'Método 2')

leg = plt.legend(loc='best', ncol=2, mode="expand", shadow=True, fancybox=True)
leg.get_frame().set_alpha(0.5)

plt.show()
```

- Dentre os dois bubble sorts, por algum motivo o primeiro foi um pouco melhor que o segundo, mas não o tornou ainda um algoritmo eficiente para muitos números, pois ainda é um algoritmo de complexidade $O(n^{2})$.

```python
def bubbleSort(A):
    for i in range(50):
        for j in range(50-1):
            if ( A[i] < A[j] ):
                aux = A[j]
                A[j] = A[i]
                A[i] = aux   
```

```python
a = [randint(1,1000) for i in range(50)]

print("A lista não ordenada:",a,"\n")
```

```python
bubbleSort(a)

print("A lista ordenada:",a)
```
